Feature extraction 

```python
dest_coordinates = df_new['World.PrimaryBattleship.Navigation.waypoints'][0]
dest_coordinates_list = ast.literal_eval(dest_coordinates)

```

```python
def calc_distance_from_dest(x,y,point):
    dx = x - point[0]
    dy = y - point[1]
    return math.sqrt(dx**2 + dy**2)

for i, point in enumerate(dest_coordinates_list):
    df_new[f'distance_to_point_{i + 1}'] = df_new.apply(lambda row: calc_distance_from_dest(row['World.PrimaryBattleship.x'], row['World.PrimaryBattleship.y'], point), axis=1)
```

```python

df_new['distance_from_airplane'] = df_new.apply(lambda row: calc_distance_from_dest(row['World.PrimaryBattleship.x'], row['World.PrimaryBattleship.y'], (row['World.AircraftCarrier.x'], row['World.AircraftCarrier.y'])), axis=1)
```

```python
df_new.columns
```

```python
df_new['World.PrimaryBattleship.RadarSonar.radar_objects'] = df_new['World.PrimaryBattleship.RadarSonar.radar_objects'].apply(ast.literal_eval)
```

```python
df_new['World.PrimaryBattleship.RadarSonar.radar_objects']
```

```python
def calculate_distance(ship_x, ship_y, radar_objects):
    distances = []
    for radar_point in radar_objects:
        if len(radar_point) > 0:
            distances.extend([np.sqrt((ship_x - x)**2 + (ship_y - y)**2) for x, y in radar_point])
    return distances[:4]

df_new[['distance_to_radar_obj_1', 'distance_to_radar_obj_2', 'distance_to_radar_obj_3', 'distance_to_radar_obj_4']] = df_new.apply(
    lambda row: pd.Series(calculate_distance(row['World.PrimaryBattleship.x'], row['World.PrimaryBattleship.y'], row['World.PrimaryBattleship.RadarSonar.radar_objects'])),
    axis=1
)
df_new[['distance_to_radar_obj_1', 'distance_to_radar_obj_2', 'distance_to_radar_obj_3', 'distance_to_radar_obj_4']] = df_new[['distance_to_radar_obj_1', 'distance_to_radar_obj_2', 'distance_to_radar_obj_3', 'distance_to_radar_obj_4']].fillna(0)
```

Training using Decision Tree

```python
#for World.PrimaryBattleship.heading
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Assuming you have a target variable 'y' for regression

# Split the data into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(df_new, y1, test_size=0.2, random_state=42)

# Create a decision tree regressor
regressor_heading = DecisionTreeRegressor()

# Fit the model on the training data
regressor_heading.fit(X_train, y_train)

# Make predictions on the test set
y_pred_heading = regressor_heading.predict(X_test)

# Calculate the mean squared error (MSE) to evaluate the model's performance
mse = mean_squared_error(y_test, y_pred_heading)
print("Mean Squared Error:", mse)
```
